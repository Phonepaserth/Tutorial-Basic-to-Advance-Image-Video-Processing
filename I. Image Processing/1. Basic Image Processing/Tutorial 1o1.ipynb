```python
import matplotlib.pyplot as plt
```

```python
X = plt.imread('frutos_rojos.png')
plt.imshow(X)
plt.show()
```

```python
# Red channel
R = X[:,:,0]
plt.imshow(R,cmap='gray')
plt.show()
```

```python
G = X[:,:,1]
plt.imshow(G,cmap='gray')
plt.show()
```
