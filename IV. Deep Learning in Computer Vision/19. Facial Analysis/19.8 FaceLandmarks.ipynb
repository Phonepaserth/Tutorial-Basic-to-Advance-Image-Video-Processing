# Face Landmarks con Dlib


## Libraries

```
import dlib
```

## Trained Model

```
!wget https://www.dropbox.com/s/l68dhjikrvd90y9/shape_predictor_68_face_landmarks.dat
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
```

## Testing Images

```
!wget https://www.dropbox.com/s/2747g2z7c4rb8ii/facesABC.zip
!unzip facesABC.zip
```

```
import cv2
import matplotlib.pyplot as plt


A = cv2.imread('faceA.png')
B = cv2.imread('faceB.png')
C = cv2.imread('faceC.jpg')

plt.imshow(cv2.cvtColor(A, cv2.COLOR_BGR2RGB))
plt.title('A')
plt.show()
plt.imshow(cv2.cvtColor(B, cv2.COLOR_BGR2RGB))
plt.title('B')
plt.show()
plt.imshow(cv2.cvtColor(C, cv2.COLOR_BGR2RGB))
plt.title('C')
plt.show()
```

## Landmark Detection

```
fimg = 'faceA.png'
img = dlib.load_rgb_image(fimg)
```

## Results

```
import numpy as np
```

```
h = img.shape[0]
w = img.shape[1]
dlib_rect = dlib.rectangle(int(0), int(0), int(w), int(h))
shape = predictor(img, dlib_rect)
x = np.zeros((68,))
y = np.zeros((68,))
for i in range(68):
  x[i] = shape.part(i).x
  y[i] = shape.part(i).y

plt.figure(figsize=(10,15))
plt.imshow(img)
plt.title(fimg)
plt.scatter(x,y)
for i in range(68):
  plt.text(x[i],y[i],str(i),c='white')
plt.show()
```
